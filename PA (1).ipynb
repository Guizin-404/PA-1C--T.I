```python
# Escrever um programa que:
# Peça o valor da compra
# Se o valor for maior que 100, aplique um desconto de 10%
# Se não, mostre o valor normal.

valor_compra = float(input("Digite o valor da compra: "))

if valor_compra > 100:
    desconto = valor_compra * 0.10
    valor_final = valor_compra - desconto
    print(f"Valor com desconto (10%): R${valor_final:.2f}")
else:
    print(f"Valor normal: R${valor_compra:.2f}")
```

```python

#Escrever um programa que realiza todas as operações aritimeticas

print("---------CALCULADORA SIMPLES--------")
print("")
print("----------------MENU----------------")
print("")
print("SOMA ----------------------------- 1")
print("")
print("SUBTRAÇÃO ------------------------ 2")
print("")
print("MULTIPLICAÇÃO -------------------- 3")
print("")
print("DIVISÃO -------------------------- 4")
print("------------------------------------")

esc = int(input("Digite a opção desejada: "))

if esc == 1:
    n1 = int(input("selecione o primeiro numero da soma: "))
    n2 = int(input("selecione o segundo numero da soma: "))
    calc = n1 + n2
    print(f"O resultado da Soma é: {calc}")

if esc == 2:
    n1 = int(input("selecione o primeiro numero da subtração:  "))
    n2 = int(input("selecione o segundo numero da subtração: "))
    calc = n1 - n2
    print(f"O resultado da Subtração é: {calc}")

if esc == 3:
    n1 = int(input("selecione o primeiro numero da multiplicação: "))
    n2 = int(input("selecione o segundo numero da multiplicação: "))
    calc = n1 * n2
    print(f"O resultad oda Multiplicação é: {calc}")

if esc == 4:
    n1 = int(input("selecione o primeiro numero da divisão: "))
    n2 = int(input("selecione o segundo numero da divisão: "))
    calc = n1 / n2
    print(f"O resultado da Divisão é: {calc}")
```

```python
# Programa que classifica os tipos de triangulos:
#isoceles
#equilarero
#Escaleno

l1 = int(input("Digite o primeiro lado do triangulo: "))
l2 = int(input("Digite o segundo lado do triangulo: "))
l3 = int(input("Digite o terceiro lado do triangulo: "))

if l1 == l2 and l2 == l3:
    print("O triangulo é equilatero")

if l1 != l2 and l2 != l3 and l1 != l3:
    print("O triangulo é escaleno")

if (l1 == l2 or l2 == l3 or l1 == l3) and (l1 != l2 or l2 != l3 or l1 != l3):
    print("O triangulo é isoceles")
```
